```python
!pip install -q laspy numpy

import laspy
import numpy as np

# Load real LiDAR data
las_path = "/content/drive/MyDrive/Photogrammetry VS LiDAR Data Sets/LiDAR Data/Urban Data/lidars/terra_las/cloud2b201cc0229f67.las"
las = laspy.read(las_path)

print(f"=== REAL LiDAR Data (DJI Drone, Urban) ===\n")
print(f"Total points: {len(las.points):,}")
print(f"X range: [{las.x.min():.2f}, {las.x.max():.2f}] ({las.x.max()-las.x.min():.1f}m)")
print(f"Y range: [{las.y.min():.2f}, {las.y.max():.2f}] ({las.y.max()-las.y.min():.1f}m)")
print(f"Z range: [{las.z.min():.2f}, {las.z.max():.2f}] ({las.z.max()-las.z.min():.1f}m)")
print(f"\nPoint format: {las.point_format}")
print(f"Available fields: {list(las.point_format.dimension_names)}")
```

```python
# Visualize it!
import matplotlib.pyplot as plt

points = np.column_stack([las.x, las.y, las.z]).astype(np.float32)

# Subsample for plotting (if too many points)
if len(points) > 50000:
    idx = np.random.choice(len(points), 50000, replace=False)
    sample = points[idx]
else:
    sample = points

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Top-down view
axes[0].scatter(sample[:,0], sample[:,1], c=sample[:,2], s=0.1, cmap='terrain')
axes[0].set_title('Top-Down View (colored by height)')
axes[0].set_xlabel('X (meters)')
axes[0].set_ylabel('Y (meters)')
axes[0].set_aspect('equal')

# Side view
axes[1].scatter(sample[:,0], sample[:,2], c=sample[:,2], s=0.1, cmap='terrain')
axes[1].set_title('Side View (colored by height)')
axes[1].set_xlabel('X (meters)')
axes[1].set_ylabel('Z (height, meters)')

plt.suptitle('REAL LiDAR Point Cloud — DJI Drone Urban Scan', fontsize=14)
plt.tight_layout()
plt.show()

print(f"\nThis is REAL data from a DJI drone!")
print(f"Your Phase 7 CUDA kernels can process this: ground removal, clustering, etc.")
```
